```python
import pandas as pd
import numpy as np

def create_constant_trace(slice_name, app_name, source, dest, protocol, 
                          length_value, n_samples=1000000, time_step=0.1):
    """Generate constant trace CSV for a slice."""
    data = {
        "Time": np.round(np.arange(0, n_samples * time_step, time_step), 6),
        "App name": [app_name] * n_samples,
        "No.": np.arange(1, n_samples + 1),
        "Source": [source] * n_samples,
        "Destination": [dest] * n_samples,
        "Protocol": [protocol] * n_samples,
        "Length": [length_value] * n_samples
    }
    df = pd.DataFrame(data)
    df.to_csv(f"{slice_name}_constant.csv", index=False)
    print(f"✅ Saved {slice_name}_constant.csv with {n_samples} samples.")

# Example usage — customize as you wish
create_constant_trace("eMBB", "YouTube", "172.30.1.1", "172.30.1.250", "TCP", length_value=1200)
create_constant_trace("mMTC", "WhatsApp", "172.30.1.1", "172.30.1.250", "TCP", length_value=200)
create_constant_trace("URLLC", "Meet", "172.30.1.1", "172.30.1.250", "UDP", length_value=600)
```
